```python
# Let's start with a list of popular travel destinations in India.
destinations = ["Hyderabad", "Madanapalle", "Rayachoti", "Punganur", "Bangalore", "Kadapa"]
print(f"Original list: {destinations}\n")
# 1. append(): Adds a single element to the end of the list.
print("--- 1. append() ---")
destinations.append("Bangalore")
print(f"After appending 'Bangalore': {destinations}\n")
# 2. extend(): Adds all elements from an iterable (like another list) to the end.
print("--- 2. extend() ---")
more_destinations = ["Kadapa", "Kanchipuram"]
destinations.extend(more_destinations)
print(f"After extending with {more_destinations}: {destinations}\n")
# 3. insert(): Inserts an element at a specified index.
print("--- 3. insert() ---")
# Let's add 'Udaipur' at index 2 (the third position).
destinations.insert(2, "Goa")
print(f"After inserting 'Goa' at index 2: {destinations}\n")
# 4. remove(): Removes the first occurrence of a specified value.
print("--- 4. remove() ---")
destinations.remove("Hyderabad")
print(f"After removing 'Hyderabad': {destinations}\n")
# 5. pop(): Removes and returns the element at a specified index.
# If no index is specified, it removes and returns the last item.
print("--- 5. pop() ---")
popped_item = destinations.pop(5) # Removes the item at index 5
print(f"Popped item at index 5: '{popped_item}'")
print(f"List after pop(5): {destinations}\n")
# 6. index(): Returns the index of the first occurrence of a specified value.
print("--- 6. index() ---")
e_index = destinations.index("Bangalore")
print(f"The index of 'Bangalore' is: {e_index}\n")
# 7. count(): Returns the number of times a specified value appears.
print("--- 7. count() ---")
destinations.append("Kadapa") # Add a duplicate for counting
print(f"List with a duplicate 'Kadapa': {destinations}")
a_count = destinations.count("Kadapa")
print(f"The count of 'Kadapa' is: {a_count}\n")
destinations.remove("Kadapa") # Clean up the duplicate
# 8. sort(): Sorts the list in place (alphabetically for strings).
print("--- 8. sort() ---")
destinations.sort()
print(f"List after sorting alphabetically: {destinations}\n")
# You can also sort in descending order.
destinations.sort(reverse=True)
print(f"List after sorting in reverse: {destinations}\n")

# 9. reverse(): Reverses the order of the elements in the list in place.
print("--- 9. reverse() ---")
destinations.reverse()
print(f"List after reversing: {destinations}\n")
# 10. copy(): Returns a shallow copy of the list.
print("--- 10. copy() ---")
destinations_copy = destinations.copy()
destinations_copy.append("M") # Modify the copy
print(f"Original list remains unchanged: {destinations}")
print(f"Copied and modified list: {destinations_copy}\n")
# 11. clear(): Removes all elements from the list.
print("--- 11. clear() ---")
destinations_copy.clear()
print(f"The copied list after clearing: {destinations_copy}")
```
