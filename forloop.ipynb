```python
# ----------------------------------------
# Python For Loop - Full Demonstration
# ----------------------------------------
print("1. For loop with a List")
fruits = ["Apple", "Banana", "Mango"]
for fruit in fruits:
    print("Fruit:", fruit)
print("\n2. For loop with a String (character by character)")
for ch in "HELLO":
    print("Character:", ch)
print("\n3. For loop with a Tuple")
numbers = (10, 20, 30)
for num in numbers:
    print("Tuple value:", num)
print("\n4. For loop with Dictionary - keys and values")
inventory = {"Apples": 5, "Bananas": 3, "Oranges": 7}
print("Only keys:")
for item in inventory:
    print(item)
print("Keys and values using items():")
for key, value in inventory.items():
    print(key, "->", value)
print("\n5. Using range()")
print("Range from 0 to 4:")
for i in range(5): # 0 to 4
    print(i)
print("Range with start and end (2 to 6):")
for i in range(2, 7):
    print(i)
print("Range with step (0 to 10, step 2):")
for i in range(0, 11, 2):
    print(i)
print("\n6. Nested for loops (Multiplication Table 1 to 3)")
for i in range(1, 4):
    for j in range(1, 4):
        print(f"{i} x {j} = {i*j}", end="\t")
    print() # new line after each row

print("\n7. For loop with else clause")
for n in range(3):
    print("Number:", n)
else:
    print("Loop finished without break!")

print("\n8. Using break inside for loop")
for n in range(5):
    if n == 3:
        print("Breaking at", n)
        break
    print("Value:", n)

print("\n9. Using continue inside for loop")
for n in range(5):
    if n == 2:
        print("Skipping at", n)
        continue
    print("Value:", n)

print("\n10. Using pass inside for loop (placeholder)")
for n in range(3):
    if n == 1:
        pass # does nothing, just a placeholder
    print("Number:", n)

print("\n11. Practical Example - Summation of numbers 1 to 10")
total = 0
for i in range(1, 11):
    total += i
print("Sum from 1 to 10 =", total)

print("\n12. Practical Example - Search in a List")
colors = ["red", "blue", "green", "yellow"]
search = "green"
for c in colors:
    if c == search:
        print(search, "found!")
        break
else:
    print(search, "not found")

# Python For Loop Demonstration - Complete Program
# -------------------------------------------------
# 1. Looping over different data types
print("1. Looping over different data types")
# List
fruits = ["Apple", "Banana", "Mango"]
print("\nList:")
for fruit in fruits:
    print(fruit)
# Tuple
numbers = (10, 20, 30)
print("\nTuple:")
for num in numbers:
    print(num)
# String
word = "HELLO"
print("\nString:")
for ch in word:
    print(ch)
# -------------------------------------------------
search = "green"
for c in colors:
    if c == search:
        print(search, "found!")
        break
    else:
        print(search, "not found")

# Dictionary
inventory = {"Apples": 5, "Bananas": 3, "Oranges": 7}
print("\nDictionary (keys and values):")
for key, value in inventory.items():
    print(key, "->", value)
# 2. Range() variations
print("\n2. Range() variations")
print("Simple range(5):")
for i in range(5): # 0 to 4
    print(i)
print("\nRange with start and end (2 to 6):")
for i in range(2, 7):
    print(i)
print("\nRange with step (0 to 10 step 2):")
for i in range(0, 11, 2):
    print(i)

# 3. Nested loops - Multiplication table
print("\n3. Nested loops - Multiplication Table (1 to 3)")
for i in range(1, 4):
    for j in range(1, 4):
        print(f"{i} x {j} = {i*j}", end="\t")
    print() # new line after each row

# 4. Else with loop
print("\n4. For loop with else")
for n in range(3):
    print("Number:", n)
else:
    print("Loop finished naturally (no break used)")

# 5. Control statements
print("\n5. Control statements")
# break
print("Using break:")
for n in range(5):
    if n == 3:
        print("Breaking at", n)
        break
    print("Value:", n)
# continue
print("\nUsing continue:")
for n in range(5):
    if n == 2:
        print("Skipping at", n)
        continue
    print("Value:", n)
# pass
print("\nUsing pass:")
for n in range(3):
    if n == 1:
        pass # placeholder (does nothing)
    print("Number:", n)

# 6. Practical Examples
print("\n6. Practical Examples")
# Sum of numbers 1 to 10
total = 0
for i in range(1, 11):
    total += i
print("Sum of numbers from 1 to 10 =", total)
# Search in a list
colors = ["red", "blue", "green", "yellow"]
search = "green"
for c in colors:
    if c == search:
        print(search, "found!")
        break
else:
    print(search, "not found")
```
