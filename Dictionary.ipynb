```python
# Create an inventory (product: quantity)
inventory = {
"Apples": 50,
"Bananas": 30,
"Oranges": 20
}
print("Initial Inventory:", inventory)

# ---- Dictionary Methods Demonstrations ----

# 1. get() - safely get item
print("\n1. Get method")
print("Apples count:", inventory.get("Apples"))
print("Mangoes count (not present):", inventory.get("Mangoes", "Not Available"))

# 2. keys() - list of keys
print("\n2. Keys method")
print("Products available:", list(inventory.keys()))

# 3. values() - list of values
print("\n3. Values method")
print("Stock quantities:", list(inventory.values()))

# 4. items() - key-value pairs
print("\n4. Items method")
print("Inventory details:", list(inventory.items()))

# 5. update() - update with another dict
print("\n5. Update method")
inventory.update({"Mangoes": 40, "Bananas": 60}) # Update Bananas, Add Mangoes
print("After update:", inventory)

# 6. copy() - create a copy
print("\n6. Copy method")
backup_inventory = inventory.copy()
print("Backup inventory:", backup_inventory)

# 7. pop() - remove specific item
print("\n7. Pop method")
removed = inventory.pop("Oranges")
print("Removed Oranges, Quantity:", removed)
print("After pop:", inventory)

# 8. popitem() - remove last inserted item
print("\n8. Popitem method")
last_removed = inventory.popitem()
print("Last removed item:", last_removed)
print("After popitem:", inventory)

# 9. setdefault() - get value if key exists, else set default
print("\n9. Setdefault method")
inventory.setdefault("Grapes", 25)
print("After setdefault:", inventory)

# 10. fromkeys() - create new dict from keys
print("\n10. Fromkeys method")
categories = ["Fruits", "Vegetables", "Dairy"]
category_stock = dict.fromkeys(categories, 0)
print("Category stock initialized:", category_stock)

# 11. clear() - remove all items
print("\n11. Clear method")
temp = inventory.copy()
temp.clear()
print("After clear:", temp)

# 12. 'in' keyword - check existence
print("\n12. 'in' keyword")
print("Is Apples available?", "Apples" in inventory)
print("Is Milk available?", "Milk" in inventory)

# 13. del - delete key
print("\n13. del keyword")
del inventory["Apples"]
print("After deleting Apples:", inventory)

# -----------------------------------
print("\nFinal Inventory:", inventory)
```
