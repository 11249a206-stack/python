Lists are ordered collections of items. You can change lists.

```python
# Example of a list
list = [1, 2, 3,"Hyderabad","Banglore"]
print(list)
```

### Sets (set)
Sets are unordered collections of unique items.

```python
# Example of a set
set = {1, 3, 5, 1 , 2 , 3}
print(set)


```

```python
# A super simple program for string operators

# --- Setup ---
first_name = "Manikyala"
last_name = " Pramodh"
word = "kohli"

# --- 1. Concatenation (+) ---
# Joining strings together
full_name = first_name + " " + last_name
print("Full Name:", full_name)

# --- 2. Repetition (*) ---
# Repeating a string
sound = word * 3
print("Meow sound:", sound)

# --- 3. Membership (in) ---
# Checking if a letter is in a word
is_a_present = "a" in full_name
print("Is 'a' in the full name?", is_a_present)
```

```python
# A super simple program for flow controls

age = 18

print("--- Checking age ---")
if age >= 18:
    print("You are an adult.")
else:
    print("You are a minor.")

```
