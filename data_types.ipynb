```python
my_integer = 10
print(f"Integer Value: {my_integer}")
print(f"Type: {type(my_integer)}\n")

# 2. Float (float):
# A number with a decimal point.
my_float = 20.5
print(f"Float Value: {my_float}")
print(f"Type: {type(my_float)}\n")

# 3. Complex (complex):
# A number with a real and imaginary part. The imaginary part is denoted by 'j'.
my_complex = 3 + 4j
print(f"Complex Number Value: {my_complex}")
print(f"Type: {type(my_complex)}\n")
```

```python
y_string = "Hello, Python!"
print(f"String Value: {my_string}")
print(f"Type: {type(my_string)}\n")

# 2. List (list):
# An ordered, mutable (changeable) collection of items.
# Lists are defined by square brackets [].
my_list = [1, "apple", 3.14, True]
print(f"List Value: {my_list}")
print(f"Type: {type(my_list)}\n")

# 3. Tuple (tuple):
# An ordered, immutable (unchangeable) collection of items.
# Tuples are defined by parentheses ().
my_tuple = (1, "banana", 6.28)
print(f"Tuple Value: {my_tuple}")
print(f"Type: {type(my_tuple)}\n")
```
