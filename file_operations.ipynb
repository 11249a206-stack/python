```python
#create file
with open("example.txt",'a')as file:
  print('file created')
```

```python
#open file
with open('example.txt','r')as file:
  print('file opened')
  # The 'with' statement automatically closes the file, so file.close() is not needed.
```

```python
#read the file
with open('example.txt','r') as file:
  content = file.read()
  print('file content : hello!',content)
```

```python
#write into the file
with open('example.txt','w') as file:
  file.write("Hello,i am your friend \n")
  file.write("This is a test file.\n")
  print('Data written to the file.')
```

```python
#save the file
with open('example.txt','a') as file:
  file.close()
  print('filE saved and closed')
```
